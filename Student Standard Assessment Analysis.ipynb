```python
# Student Assessment Data Visualizations
# Skills used: data visualization (donut and bar charts), data aggregation, data manipulation, conditional coloring, color mapping, and error handling


import pandas as pd
import matplotlib.pyplot as plt 

# Using pandas to analyze data

# Reading the database
data = pd.read_csv("Sample Student Standards Data.csv")
data.head()
df = pd.DataFrame(data)

# Printing the top 10 rows
display(data.head(10))
```

```python
# Calculating data in Domain 1
d1_score_counts = data['Domain 1'].value_counts()

# Create a custom color map in Domain 1
colors = {'M': 'steelblue', 'P': 'green', 'B': 'orange', 'IE': 'brown'}
pie_colors1 = [colors[domain] for domain in d1_score_counts.index]

# Creating a donut chart to show assessment results in Domain 1
fig, ax = plt.subplots()
ax.pie(d1_score_counts, labels=d1_score_counts.index, autopct='%1.1f%%', startangle=90, pctdistance=0.85, colors=pie_colors1)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

ax.axis('equal')
plt.title('Proportion of each Score in Domain 1')
plt.show()
```

```python
# Calculating data in two specified columns: IEP and Ethnicity

iep_ethnicity_counts = data[data['IEP'] == 'Yes'].groupby('Ethnicity').size()

#Creating a bar plot to visualize IEPs by Ethnicity

iep_ethnicity_counts.plot(kind='bar')
plt.title('Student IEP and Ethnicity Breakdown')
plt.xlabel('Ethnicity')
plt.ylabel('Count')
plt.show()


```
